```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/multiple_linear_regression.csv')
```

```python
data.head()
```

```python
data.info()
```

## Data Preprocessing

```python
data = data.dropna()
data.shape
```

## splitting data

```python
output_col = "Target"
```

```python
x = data.iloc[:, data.columns != output_col]
```

```python
x.head()
```

```python
y = data.loc[:, output_col]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size = 0.30, random_state = 43)
```

```python
x_train
```

#Linear Regression


##Training model


```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

##Predicting Test_input


```python
predicted_value = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

cost = mean_squared_error(y_test, predicted_value)
```

```python
cost
```

```python
plt.plot(x_test, y_test, "*", color = "green")
plt.plot(x_test, predicted_value, "+", color = "red")
plt.title("Performance testing")
plt.xlabel("Input")
plt.xticks(np.linspace(1, 500, 10))
plt.ylabel("Output")
plt.show()
```

## Assumptions

### Homoscedasticity

```python
predicted_value
```

```python
residuals = y_test - predicted_value
```

```python
residuals
```

```python
plt.scatter(predicted_value, residuals)
plt.xlabel("y_pred value")
plt.ylabel("residual values")
plt.show
```

## Ordinary Least Square (OLS) Method

```python
import statsmodels.api as sm
```

```python
data.head()
```

```python
x_with_constant = sm.add_constant(x_train)
x_with_constant
```

```python
model = sm.OLS(y_train, x_with_constant)
```

```python
result = model.fit()
```

```python
result.params
```
